# Car Flipping

**Contexto:** El car flipping es un modelo de negocio que consiste en la compra de vehiculos que requieran algun tipo de reparacion, relizarlas, y obtener una ganancia al vender el vehiculo ya reparado y en condiciones.

**Luxus Revive** es una empresa que desde el 2005 se dedica al car flipping de vehiculos exclusivos y de alta gama.
La empresa realiza las compras mayoritariamente en subastas, priorizando un millage bajo, ya que vehiculos con mayor cantidad de millas recorridas son mas dificiles de vender.

Debido al crecimiento de la empresa y al surgimiento de nuevos competidores **Luxus Revive** busca, utilizando los datos, abrir nuevos consesionarios, maximizar sus ganacias y optimizar los viajes/visitas que se realizan para evaluar cada vehiculo previo a la compra.

**Intereses de la investigacion:** El objetivo de esta investigacion es encontrar las mejores marcas y selecionar los menores kilometrajes en vehiculos, analizar posibles ubicaciones para la apertura de nuevas sucursales, teniendo en cuenta los estados donde mas vehiculos de alta gama se encuentran a la venta.

# Clusterizacion para nuevas sucursales

```python
X = df[['price', 'mileage', 'state']]
```

```python
X_encoded = pd.get_dummies(X, columns=['state'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

```python
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

```python
df['cluster'] = clusters
```

# Grafico del clustering

```python
cluster_colors = ['red', 'darkviolet', 'lightgreen']
```

```python
cluster_names = ['NY, MI, IL, WA, GA', 'TX', 'MN, CA, WI, FL, NJ, TN']
```

```python

plt.scatter(df['price'], df['mileage'], c=df['cluster'], cmap='rainbow')
legend_elements = [plt.Line2D([0], [0], marker='o',color='w', markerfacecolor=cluster_colors[i], markersize=10, label=cluster_names[i]) for i in range(n_clusters)]

plt.legend(handles=legend_elements)
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.title('Subastas de Vehiculos por Estado')
plt.show()
```

```python
conteo_cluster = df['cluster'].value_counts()
conteo_cluster
```
